```python
# Remove all Unsampled Rows
df_filtered = df[df['Sample_taken'] == 'Sampled']
df_filtered.shape
```

```python
# Remove Unnecessary Features
df_filtered = df_filtered.drop(columns=['WP_ID','DataType','Date_Assessment_Original','SURVEY_DETAIL_ID','COUNTRY','Comment','HCO3','Corrected_HCO3','Sample_taken'],axis=1)
df_filtered.shape
```

```python
df_filtered.sample(5)
```

```python
# Check for Missing Values
df_filtered.isnull().sum()
```

```python
# Impute Missing Values
from sklearn.impute import KNNImputer

df_drop = df_filtered.drop(columns=['Date_Assessment','Time_Assessment'])

knn_imputer = KNNImputer()
df_imputed = knn_imputer.fit_transform(df_drop)

# Convert the imputed array back to a DataFrame
df_knn = pd.DataFrame(df_imputed, columns=df_drop.columns)

# Check for missing values
df_knn.isnull().sum()
```

```python
df_drop.sample(5)
```

```python
# Display Imputed Dataset
df_knn.sample(5)
```

```python
# Display the Distribution

sns.set(style="whitegrid")

# Define colors for the plots
# colors = sns.color_palette("pastel")

num_columns = len(df_knn.columns)
num_plots_per_row = 3
num_rows = (num_columns + num_plots_per_row - 1) // num_plots_per_row  # Calculate the number of rows needed

# Adjust figure size
fig, axes = plt.subplots(num_rows, num_plots_per_row, figsize=(15, 3*num_rows))

# Flatten the axes array if it's not already 1-dimensional
axes = axes.flatten() if num_rows > 1 else [axes]

# Iterate through each column and plot its distribution
for i, column in enumerate(df_knn.columns):
    sns.histplot(df_knn[column], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {column}', fontsize=14)
    axes[i].set_xlabel(column, fontsize=12)
    axes[i].set_ylabel('Frequency', fontsize=12)
    axes[i].tick_params(axis='both', which='major', labelsize=10)

# Remove empty subplots
for j in range(num_columns, num_rows * num_plots_per_row):
    fig.delaxes(axes[j])

plt.tight_layout()
# plt.savefig('plot1.png')
plt.show()
```

```python
# Normal Distribution Check
from tabulate import tabulate
from scipy.stats import shapiro

results = []

for column in df_knn.columns:
    stat, p = shapiro(df_knn[column])
    if p > 0.05:
        results.append([column, "Normally Distributed", p])
    else:
        results.append([column, "Not Normally Distributed", p])

print(tabulate(results, headers=["Column", "Distribution", "p-value"], tablefmt="grid"))

# # Convert the results list into a DataFrame
# results_df = pd.DataFrame(results, columns=["Column", "Distribution", "p-value"])

# # Display the DataFrame
# results_df
```

```python
# Box_Plot for Outliers Detection

sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))

# Create a boxplot for each column in df_knn_cont
sns.boxplot(data=df_knn, palette="Set3", fliersize=5, linewidth=1.5)

# Add title and labels
plt.title('Boxplot of Data with Outliers', fontsize=16)
plt.xlabel('Columns', fontsize=14)
plt.ylabel('Values', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
sns.despine()

plt.grid(True, linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
# plt.savefig('boxplot1.png')
plt.show()
```

```python
# Outlier Detection and Treatment

df_knn_rest = df_knn.copy()
df_s = np.sort(df_knn, axis=0)

# Calculate Quartiles and IQR for all columns
Q1 = np.percentile(df_s, 25, axis=0)
Q3 = np.percentile(df_s, 75, axis=0)
IQR = Q3 - Q1

# Compute lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Original indices
original_indices = df_knn_rest.index

# Identify and replace outliers for each column
outliers = np.any((df_s < lower_bound) | (df_s > upper_bound), axis=0)
for i, column in enumerate(df_s.T):
    column_outliers = (column < lower_bound[i]) | (column > upper_bound[i])
    if np.any(column_outliers):
        # Replace outliers with median
        df_s[column_outliers, i] = np.median(column)
        # Restore the original order of rows
        df_s[:, i] = df_s[original_indices, i]

# Check if any column still contains outliers
outliers_after_replacement = np.any((df_s < lower_bound) | (df_s > upper_bound), axis=0)

# Print the result
if np.any(outliers_after_replacement):
    print("Some columns still contain outliers after replacement.")
    print("Columns with outliers:")
    for i, column in enumerate(outliers_after_replacement):
        if column:
            print(f"Column {i+1}")
else:
    print("No columns contain outliers after replacement.")
```
